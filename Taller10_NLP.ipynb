```python
import pandas as pd
```

```python
import numpy as np
```

```python
from wordcloud import WordCloud, ImageColorGenerator
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("spanish")
```

```python
import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

```python
data = pd.read_csv("C:/Users/AndresFelipe/Downloads/CaperucitaRoja.csv", sep=";", error_bad_lines=False)
data.head()
```

```python
data = data.drop(['Unnamed: 2'], axis=1)
```

```python
data = data.drop([4, 5],axis=0)
data.head()
```

```python
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          colormap="Dark2",
                          random_state=42
                         ).generate(str(data['Caperucita Roja']))

print(wordcloud)
plt.figure(figsize = (10,8))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
